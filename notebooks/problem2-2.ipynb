```python
import networkx as nx
import matplotlib.pyplot as plt

# ネットワークジェネレータを初期化
generator = LatticeGraph(20, 20, 100)

# グラフを生成
G = generator.generate()

# ノードの位置を設定（ノードの座標をそのまま使用）
pos = {node: node for node in G.nodes()}

# ノードとエッジを描画
nx.draw_networkx_nodes(G, pos, node_size=30, node_color='blue', alpha=1)
nx.draw_networkx_edges(G, pos, edge_color='black', width=1)
plt.axis('off')
plt.show()
```
